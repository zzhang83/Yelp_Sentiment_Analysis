```python
test = [[r3['text'][i],r3['stars'][i]] for i in range(len(r3))]
```

```python
import pandas as pd
import numpy as np
from nltk.stem.wordnet import WordNetLemmatizer
from nltk.stem.porter import *
from nltk.stem.lancaster import LancasterStemmer
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import pprint
import string
import math
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import BernoulliNB

import nltk
# stemmer = PorterStemmer()
# lmtzr = WordNetLemmatizer()
stemmer = nltk.stem.SnowballStemmer('english')

yelp_pos = []
yelp_neg = []
amazon_pos = []
amazon_neg = []
imdb_pos = []
imdb_neg = []

with open("yelp_labelled.txt","r") as f:
    for line in f.readlines():
        l = line.strip().split('\t')
        if l[1] == '1':
            yelp_pos.append(l)
        else:
            yelp_neg.append(l)

with open("amazon_cells_labelled.txt","r") as f:
    for line in f.readlines():
        l = line.strip().split('\t')
        if l[1] == '1':
            amazon_pos.append(l)
        else:
            amazon_neg.append(l)

with open("imdb_labelled.txt","r") as f:
    for line in f.readlines():
        l = line.strip().split('\t')
        if l[1] == '1':
            imdb_pos.append(l)
        else:
            imdb_neg.append(l)

data_train = yelp_pos + yelp_neg  # + amazon_pos +amazon_neg + imdb_pos + imdb_neg
data_train[0]

def change(x):
    if x[1] >= 3.0:
        x[1] = 1
    else:
        x[1] = 0
[change(x) for x in test]

for i in data_train:
    line = i[0]
    line = line.lower() # lower case
    line = line.translate(string.punctuation)
    line = line.strip()
    split = word_tokenize(line)
    # lemmatize
    #stem and remove stopwords
    words = (set(split)) - set((stopwords.words('english')))
    data = [stemmer.stem(word) for word in words]   # reducing words to their word stem
    i[0] = data
    

for i in test:
    line = i[0]
    line = line.lower() # lower case
    line = line.translate(string.punctuation)
    line = line.strip()
    split = word_tokenize(line)
    # lemmatize
    #stem and remove stopwords
    words = (set(split)) - set((stopwords.words('english')))
    data = [stemmer.stem(word) for word in words]   # reducing words to their word stem
    i[0] = data
    
import copy
dtrain = copy.deepcopy(data_train)
dtest = copy.deepcopy(test)

#find unique words
unique = set()
for i in data_train:
    for word in i[0]:
        unique.add(word)

wordlist = list(unique)

#construct and normalize the feature vectors
for i in data_train:
    array = np.zeros(len(wordlist))
    for j in i[0]:
        array[wordlist.index(j)] += 1
    i[0] = [math.log(x+1) for x in array]
    i[1] = int(i[1]) # i[1] contains an char which is 0 or 1, need to convert to int
    
for i in test:
    array = np.zeros(len(wordlist))
    for j in i[0]:
        if j in wordlist:
            array[wordlist.index(j)] += 1
    i[0] = [math.log(x+1) for x in array]
    i[1] = int(i[1])


#logistic regression
f_train = pd.DataFrame(data_train,columns=['data','label'])
f_test = pd.DataFrame(test,columns=['data','label'])

train_data = list(f_train['data'])
train_label = list(f_train['label'])

test_data = list(f_test['data'])
test_label = list(f_test['label'])

log = LogisticRegression()
log.fit(train_data,train_label)
print(log.score(test_data,test_label))


#confusion matrix
from sklearn.metrics import confusion_matrix
y_true = log.predict(test_data)
print(confusion_matrix(test_label, y_true))

#weight vector
import heapq
nmax = heapq.nlargest(10, log.coef_[0])

print ("top 10 positive words")
for i in nmax:
    print (wordlist[np.where(log.coef_[0] == i)[0][0]])
    
nmin = heapq.nsmallest(10, log.coef_[0])
print ("top 10 negative words")
for i in nmin:
    print (wordlist[np.where(log.coef_[0] == i)[0][0]])
```

```python
#2-gram model
#construct 2-gram model
#find unique 2 grams
unique = set()
for i in dtrain:
    for j in range(len(i[0])-1):
        pair = i[0][j] + " " + i[0][j+1]
        unique.add(pair)

wordlist = list(unique)

#construct and normalize the feature vectors
for i in dtrain:
    array = np.zeros(len(wordlist))
    for j in range(len(i[0])-1):
        pair = i[0][j] + " " + i[0][j+1]
        array[wordlist.index(pair)] += 1
    i[0] = [math.log(x+1) for x in array]
    i[1] = int(i[1])
    
for i in dtest:
    array = np.zeros(len(wordlist))
    for j in range(len(i[0])-1):
        pair = i[0][j] + " " + i[0][j+1]
        if pair in wordlist:
            array[wordlist.index(pair)] += 1
    i[0] = [math.log(x+1) for x in array]
    i[1] = int(i[1])
```
